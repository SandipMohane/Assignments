```python
df = df.drop(columns=['state'],axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x = sc.fit_transform(df)

x
```

```python
x = pd.DataFrame(x,columns=df.columns)
x
```

# HC_clustering 

```python
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN
```

```python
lk_complete = hierarchy.linkage(x,method='complete')
dendrogram_complete = hierarchy.dendrogram(lk_complete)
plt.axhline(y = 5 , c = 'violet', linestyle = '--')
```

```python
# model preparation for Hc

AC = AgglomerativeClustering()

Y_HC = AC.fit_predict(x)

Y_HC
```

```python
df['Hc_cluster'] = Y_HC

df.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, silhouette_score
```

```python
silhouette_score(x,Y_HC)
```

# kmeans Clustering 

```python
 from sklearn.cluster import KMeans
```

```python
km = KMeans()

kmean=km.fit_predict(x)

kmean
```

```python
# elbow Method 

wcss = []


for i in range(1,30):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,30),wcss)
plt.xlabel('K_values')
plt.ylabel('WCSS')
```

```python
km = KMeans(n_clusters=4)

ylabel=km.fit_predict(x)

ylabel
```

```python
km.cluster_centers_
```

```python
df['K_mean_cluster']  = ylabel

df
```

```python
silhouette_score(x,K_mean_cluster)
```

# DBSCAN

```python
from sklearn.neighbors import NearestNeighbors


neigh = NearestNeighbors(n_neighbors = 5).fit(x)
dist, ind = neigh.kneighbors(x)
```

```python
pd.DataFrame(dist)
```

```python
distance = np.sort(dist[:,1], axis = 0)

plt.plot(distance)
plt.xlabel('Number of samples')
plt.ylabel('Epsilon Value')
plt.title('K-dist Plot')
```

```python
db = DBSCAN(eps = 1.15, min_samples = 3)
y_db = db.fit_predict(x)

y_db
```

```python
df['Db_cluster'] = y_db

df.head()
```

```python
silhouette_score(x,y_db)
```
